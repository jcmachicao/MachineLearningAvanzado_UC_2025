```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
```

```python
# 1. Cargar un dataset de juguete (tweets en español con etiquetas de sentimiento)
dataset = load_dataset("pysentimiento/spanish-tweets")

# 2. Cargar el tokenizer y el modelo preentrenado
model_name = "dccuchile/bert-base-spanish-wwm-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)
```

```python
# 3. Tokenizar el dataset
def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True)
dataset = dataset.map(tokenize, batched=True)

# 4. Configurar entrenamiento (solo unas pocas iteraciones para demo)
args = TrainingArguments(
    output_dir="results",
    evaluation_strategy="epoch",
    save_strategy="no",
    per_device_train_batch_size=8,
    num_train_epochs=1,  # solo una pasada rápida
    logging_steps=10,
)

# 5. Definir entrenador
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=dataset["train"].shuffle().select(range(200)),  # subset pequeño
    eval_dataset=dataset["test"].shuffle().select(range(50)),
)
```

```python
# 6. Entrenar (fine-tuning ligero)
trainer.train()

# 7. Probar en una frase nueva
pred = trainer.predict(dataset["test"].select(range(1)))
print(pred)
```
